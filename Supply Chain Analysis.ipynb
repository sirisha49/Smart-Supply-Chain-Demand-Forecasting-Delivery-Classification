## Model Training

```python
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.metrics import accuracy_score, classification_report
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB
```

```python
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from xgboost import XGBClassifier
from sklearn.ensemble import AdaBoostClassifier
import numpy as np

models = {
    "SVM": (SVC(), {
        "C": [0.1, 0.001, 1, 10],
        "kernel": ["linear", "rbf", "poly"],
        "gamma": ["scale", "auto", 0.01, 0.1, 0.001],
        "degree": [2, 3, 4, 7, 9],  # For polynomial kernel
    }),
    "KNeighbors": (KNeighborsClassifier(), {
        "n_neighbors": [2, 3, 5, 7, 9],
        "weights": ["uniform", "distance"],
        "metric": ["euclidean", "manhattan", "minkowski"],
        "p": [1, 2],  # Manhattan (p=1), Euclidean (p=2)
    }),
    "NaiveBayes": (GaussianNB(), {
        "var_smoothing": np.logspace(0, -9, num=10),  # Explore smoothing parameters
    }),
    "XGBoost": (XGBClassifier(use_label_encoder=False, eval_metric='mlogloss'), {
        "n_estimators": [50, 100, 150],
        "learning_rate": [0.01, 0.1, 0.2],
        "max_depth": [3, 5, 7],
        "subsample": [0.8, 0.6],
        "colsample_bytree": [0.8, 0.6],
    }),
    "AdaBoost": (AdaBoostClassifier(), {
        "n_estimators": [50, 100, 150],
        "learning_rate": [0.01, 0.1, 1],
        "algorithm": ['SAMME', 'SAMME.R']
    })
}
```

```python
from sklearn.model_selection import RandomizedSearchCV
import numpy as np

best_model = None
best_test_score = 0
best_name = ""

for name, (model, param_grid) in models.items():
    print(f"Optimizing {name}...")
    # Use RandomizedSearchCV for hyperparameter optimization
    random_search = RandomizedSearchCV(model, param_grid, n_iter=10,  # You can adjust n_iter to your needs
                                       scoring='accuracy', cv=5, random_state=42, verbose=3)
    random_search.fit(x_train, y_train)

    # Evaluate the model on the test set
    test_score = random_search.best_estimator_.score(x_test, y_test)  # Testing accuracy

    print(f"Best parameters for {name}: {random_search.best_params_}")
    print(f"CV accuracy for {name}: {random_search.best_score_}")
    print(f"Test accuracy for {name}: {test_score}\n")

    # Track the model with the best test accuracy
    if test_score > best_test_score and test_score < 0.98:
        best_test_score = test_score
        best_model = random_search.best_estimator_
        best_name = name

# Summary of the best model
print(f"The best model is {best_name} with a test accuracy of {best_test_score}.")
```

```python
# Final report on the best model
print(f"Best Model: {best_name}")
print(f"Best Test Set Accuracy: {best_test_score}")
y_pred = best_model.predict(x_test)
print("Classification Report:")
print(classification_report(y_test, y_pred))

```
